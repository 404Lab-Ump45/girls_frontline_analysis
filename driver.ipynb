```python
import util
import matplotlib.pyplot as plt
```

```python
agents = [[util.yester, 40],         # 昨天人，永远投昨天最多的
          [util.anti_yester, 30],    # 反昨天人，永远投昨天最少的
          [util.randomer, 20],       # 随机人，随便投
          [util.rational, 10],       # 理性人，投过去最多的
          [util.learner, 0]           # 机器学习人，试图找到[历史数据]->[下一天最佳选择]
         ]
DAYS = 15
```

```python
state = util.State()
```

```python
for day in range(0, DAYS):   
    votes = [0, 0, 0]
    for i in range(0, len(agents)):
        agent, pop = agents[i]
        if pop == 0:
            continue
        votes[agent(state)] += pop
    state.update(votes)
```

```python
fig = plt.figure()
fig.suptitle('Days and Rewards')
x = range(0, DAYS)
y1 = [i[0] for i in state.history_rewards[1:]]
y2 = [i[1] for i in state.history_rewards[1:]]
y3 = [i[2] for i in state.history_rewards[1:]]
plt.plot(x, y1, label = '30')
plt.plot(x, y2, label = '40')
plt.plot(x, y3, label = '50')
plt.legend()
plt.show()
```
